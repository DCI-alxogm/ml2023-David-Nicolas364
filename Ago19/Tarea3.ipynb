```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Datos de salarios dependiendo de posición
dataset = pd.read_csv('Position_Salaries.csv')
dataset
x = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
x= x.reshape(10,)
```

```python
plt.scatter(x,y)
plt.xlabel('Nivel de posición')
plt.ylabel('Salario')
```

```python
#Para X lineal
def w(x,y):
  X=np.vstack((np.ones_like(x),x)).T
  XT=X.T
  inv= np.linalg.inv(np.matmul(XT,X))
  t=y
  w=np.matmul(inv,np.matmul(XT,t))
  return w
```

```python
M=w(x,y)
```

```python
Y=M[1]*x+M[0]
```

```python
plt.plot(x,Y)
plt.scatter(x,y)
plt.xlabel('Nivel de posición')
plt.ylabel('Salario')
plt.title("Usando X")
```

```python
#Para X cuadrático
#Se crea la matriz X, se calcula la transversa
#Se multiplica con la matriz X original y le sacamos su inversa
# Se multiplica la matriz inversa con la multiplicación de X transversa con t(valores y), y obtenemos los pesos
def w1(x,y):
  X=np.vstack((np.ones_like(x),x, x**2)).T
  XT=X.T
  inv= np.linalg.inv(np.matmul(XT,X))
  t=y
  w=np.matmul(inv,np.matmul(XT,t))
  return w
```

```python
M1 = w1(x,y)
```

```python
Y1=M1[0]+M1[1]*x+M1[2]*x**2
```

```python
plt.plot(x,Y1)
plt.scatter(x,y)
plt.xlabel('Nivel de posición')
plt.ylabel('Salario')
plt.title("Usando X^2")
```

```python
#Para X cúbico se le agregará x**3 para agregar otra potencia a la función
def w2(x,y):
  X=np.vstack((np.ones_like(x),x, x**2, x**3)).T
  XT=X.T
  inv= np.linalg.inv(np.matmul(XT,X))
  t=y
  w=np.matmul(inv,np.matmul(XT,t))
  return w
```

```python
M2 = w2(x,y)
```

```python
Y2=M2[0]+M2[1]*x+M2[2]*x**2+M2[3]*x**3
```

```python
plt.plot(x,Y2)
plt.scatter(x,y)
plt.xlabel('Nivel de posición')
plt.ylabel('Salario')
plt.title("Usando X^3")
```

```python
#Para X a la cuarta se le agregará x**4 para agregar otra potencia a la función
def w3(x,y):
  X=np.vstack((np.ones_like(x),x, x**2, x**3, x**4)).T
  XT=X.T
  inv= np.linalg.inv(np.matmul(XT,X))
  t=y
  w=np.matmul(inv,np.matmul(XT,t))
  return w
```

```python
M3 = w3(x,y)
```

```python
Y3=M3[0]+M3[1]*x+M3[2]*x**2+M3[3]*x**3+M3[4]*x**4
```

```python
plt.plot(x,Y3)
plt.scatter(x,y)
plt.xlabel('Nivel de posición')
plt.ylabel('Salario')
plt.title("Usando X^4")
```
