```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.loadtxt("olympic100m_men.txt",delimiter=',').T
```

```python
x=data[0]
y=data[1]
```

```python
plt.scatter(x,y)
```

```python
XY = x*y
N = len(x)
X2 = x**2
m = (np.sum(XY)-(np.sum(x)*np.sum(y))/N)/(np.sum(X2)-((np.sum(x))**2)/N)
b = (np.sum(y)-m*np.sum(x))/N
```

```python
m, b
```

```python
Y=m*x+b
```

```python
plt.plot(x,Y)
plt.scatter(x,y)
```
