```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Datos de mediciones de la presión sanguínea sistólica (medida en mm de Mercurio) para 29 sujetos de diferentes edades
datos = pd.read_csv('dataset.csv', sep=",", skiprows=32, usecols=[2,3])

x1 = datos['Age'].values
y1 = datos['Systolic blood pressure'].values
```

```python
plt.scatter(x1,y1)
```

```python
XY1 = x1*y1
N1 = len(x1)
X22 = x1**2
m1 = (np.sum(XY1)-(np.sum(x1)*np.sum(y1))/N1)/(np.sum(X22)-((np.sum(x1))**2)/N1)
b1 = (np.sum(y1)-m1*np.sum(x1))/N1
```

```python
m1,b1
```

```python
Y1=m1*x1+b1
```

```python
plt.plot(x1,Y1)
plt.scatter(x1,y1)
```
